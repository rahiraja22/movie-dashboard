```python
import pandas as pd
import numpy as np

# Example: Your Voting_Count column is a string with things like '(1.3K)'
def convert_votes(value):
   # if value is not np.nan:
    value = str(value)
    value = value.replace('(', '').replace(')', '').replace(',', '').strip()
    if 'K' in value:
        return int(float(value.replace('K', '')) * 1000)
    elif value.isdigit():
        return int(value)
    else:
        return 0  # fallback for unexpected values

# Apply the conversion
data['Voting_Count'] = data['Voting_Count'].apply(convert_votes)

# Check result
print(data['Voting_Count'].head())
```

```python
import pandas as pd
import re

# Function to convert "3h 1m" or "59m" into total minutes
def convert_duration(duration):
    duration = str(duration).strip()

    # Regex match
    hours = re.search(r'(\d+)h', duration)
    minutes = re.search(r'(\d+)m', duration)

    total_minutes = 0
    if hours:
        total_minutes += int(hours.group(1)) * 60
    if minutes:
        total_minutes += int(minutes.group(1))

    # If neither hours nor minutes found (e.g., "TV-MA"), return 0 or None
    return total_minutes if total_minutes > 0 else 0

# Apply to your DataFrame
data['Duration'] = data['Duration'].apply(convert_duration)

# Check result
print(data[['movie_name', 'Duration']].head())
```
